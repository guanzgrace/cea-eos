```python
import matplotlib.pyplot as plt
import matplotlib.ticker as tkr
import pandas as pd
import numpy as np
```

```python
plt.rcParams['font.size'] = 18
plt.rcParams['font.family'] = 'Arial'
plt.rc('xtick', labelsize='x-small')
plt.rc('ytick', labelsize='x-small')
```

```python
cra_color = 'darkorange'
nsc_color = 'dodgerblue'
eco_color = 'dimgrey'
```

```python
x1 = np.arange(0.0, 860, 860-0-1) #Cost of implementing NSC per infant
y1 = x1 + 220
x2 = np.arange(860, 2001, 2001-860-1)
y2 = 0*x2 + 1080
y3 = np.arange(1080, 2001, 2001-1080-1)
x3 = 0*y3 + 860
```

```python
fig, ax = plt.subplots(figsize=(8,6))
ax.get_yaxis().set_major_formatter(
    tkr.FuncFormatter(lambda x, p: format(int(x), ',')))
ax.get_xaxis().set_major_formatter(
    tkr.FuncFormatter(lambda x, p: format(int(x), ',')))
ax.set(xlabel='Cost of Implementing NSC Per Infant ($)', ylabel='Cost of Implementing ECO Per Infant ($)')
plt.plot(x1, y1, color='white',label='Switch')
plt.plot(x2, y2, color='white',label='Switch')
plt.plot(x3, y3, color='white',label='Switch')
ax.fill_between(x1, 0, y1, color=eco_color)#, alpha=0.3)
ax.fill_between(x2, 0, y2, color=eco_color)#, alpha=0.3)
ax.fill_between(x1, y1, 2000, color=nsc_color)#, alpha=0.3)
ax.fill_between(x2, y2, 2000, color=cra_color)#, alpha=0.3)
plt.text(1050, 500, "ECO", color='white', weight='extra bold')
plt.text(300, 1200, "NSC", color='white', weight='extra bold')
plt.text(1300, 1500, "CRA", color='white', weight='extra bold')
plt.ylim(-100, 2100)
plt.grid(True, linestyle='--', color='gainsboro')
plt.show()
```
