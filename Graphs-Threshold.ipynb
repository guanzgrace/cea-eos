```python
import matplotlib.pyplot as plt
import matplotlib.ticker as tkr
import pandas as pd
import numpy as np
```

```python
plt.rcParams['font.size'] = 18
plt.rcParams['font.family'] = 'Arial'
plt.rc('xtick', labelsize='x-small')
plt.rc('ytick', labelsize='x-small')
```

```python
cra_color = 'darkorange'
nsc_color = 'dodgerblue'
eco_color = 'dimgrey'
```

```python
c_nsc = 420
c_eco = 480
c_diff = c_eco - c_nsc

x1 = np.arange(0.0, c_nsc, c_nsc-0-1) #Cost of implementing NSC per infant
y1 = x1 + c_diff
x2 = np.arange(c_nsc, 1001, 1001-c_nsc-1)
y2 = 0*x2 + c_eco
y3 = np.arange(c_eco, 1001, 1001-c_eco-1)
x3 = 0*y3 + c_nsc
```

```python
fig, ax = plt.subplots(figsize=(8,6))
ax.get_yaxis().set_major_formatter(
    tkr.FuncFormatter(lambda x, p: format(int(x), ',')))
ax.get_xaxis().set_major_formatter(
    tkr.FuncFormatter(lambda x, p: format(int(x), ',')))
ax.set(xlabel='Cost of Implementing NSC Per Infant ($)', ylabel='Cost of Implementing ECO Per Infant ($)')
plt.plot(x1, y1, color='white',label='Switch')
plt.plot(x2, y2, color='white',label='Switch')
plt.plot(x3, y3, color='white',label='Switch')
ax.fill_between(x1, 0, y1, color=eco_color)#, alpha=0.3)
ax.fill_between(x2, 0, y2, color=eco_color)#, alpha=0.3)
ax.fill_between(x1, y1, 1000, color=nsc_color)#, alpha=0.3)
ax.fill_between(x2, y2, 1000, color=cra_color)#, alpha=0.3)
plt.text(550, 200, "ECO", color='white', weight='extra bold')
plt.text(150, 550, "NSC", color='white', weight='extra bold')
plt.text(650, 700, "CRA", color='white', weight='extra bold')
#plt.ylim(-100, 1100)
plt.grid(True, linestyle='--', color='gainsboro')
plt.show()
```
